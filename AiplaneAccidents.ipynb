```python
import multiprocessing
import pandas as pd
import numpy as np

def _apply_df(args):
    df, total, kwargs = args
    axis =kwargs.pop('axis')
    return df.apply(total, axis=1)

def apply_by_multiprocessing(df, total, **kwargs):
    workers = kwargs.pop('workers')
    
    pool = multiprocessing.Pool(processes=workers)
    results=pool.map(_apply_df, [(d, total, kwargs)
            for d in np.array_split(df, workers)])
    pool.close()
    
    return pd.concat(results)
    

    

monthly_injurie=apply_by_multiprocessing(av, total, axis=1, workers=4) 
print(type(monthly_injurie))
```

** I applied multiprocessing to the "total" function. I have used wrapper functions to first create a pool of threads and assign them blocks of the pd, "av" to process.I have parallized the APPLY() in-built functionality of pandas. One thing to note here is though the apply function is faster than explicit lines of code it still has a constant time for every operation. So simply using APPLY on all rows will time-out. Therefore each core  needed for parallization. Pool function of multiprocessing module has in-built support for checking number of cores available and then utilizing them**
